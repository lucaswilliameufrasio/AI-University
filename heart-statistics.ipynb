```python
import pandas
# load heart.csv
data = pandas.read_csv('heart.csv')
```

```python
print("\033[1m Média \033[0m")
print(data.mean())

print("\n\033[1m Mediana \033[0m")
print(data.median())

print("\n\033[1m Percentil \033[0m")
print(data.quantile(0.15))

print("\n\033[1m Intervalos \033[0m")
print(data.max(numeric_only=True) - data.min(numeric_only=True))

print("\n\033[1m Variância \033[0m")
print(data.var())

print("\n\033[1m Desvio Padrão \033[0m")
print(data.std())
```

```python
data.dropna().describe()
```

```python
from matplotlib import pyplot

color_class = {
    0 : 'red',
    1 : 'blue'
}

colors = [color_class[sex] for sex in data['sex']]

data.plot(kind='scatter', x='age', y='trestbps', c=colors)

pyplot.show()
```
